## Data Creation
Equation: y=10x

Target-->

x={0,1,2,3,4,5,.....}

y={0,10,20,30,40,50,....}

```python
x=[i for i in range(21)]
y=[i for i in range(201) if i%10==0]
```

## Approach 1

Defining a function for the Equation : y=10x

```python
def tempFunc(x):
  y=10*x
  return y

for value in x:
  print(tempFunc(value))
```

## Approach 2

I am super enthusiastic about ML. So i am gonna make a redundant Ml model for the Table 10

You are going to need xTrain, yTrain, xTest ,yTest from x and y lists

```python

xTrain=x[:-5] # Training data
yTrain=y[:-5] # Training Labels

xTest=x[-5:] # Testing Data
yTest=y[-5:] # Testing Labels
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# perceptron model
model = tf.keras.Sequential([
                           tf.keras.layers.Dense(units=1,input_shape=[1])     # unit is the number of neurons and input shape is the shape of the  input that you gave to neurons
])

```

```python
model.compile(optimizer='adam',loss='mae') #optimizer--> sgd,adammax,adamgrad,rmsprop; loss--> mse, mae
```

```python
model.fit(x=xTrain,y=yTrain,validation_data=(xTest,yTest), epochs=10000)
```
